```python
import pandas as pd
import numpy as np
dates=pd.date_range("20201222",periods=6)
a=pd.DataFrame(np.arange(24).reshape((6,4)),index=dates,columns=["A","B","C","D"])
print(a)
print("***********")
a.iloc[0,1]=np.nan
a.iloc[1,2]=np.nan
print(a)
print("***********")
print(a.dropna(axis=0,how="any"))         ##屏蔽缺失数据的一行
print("***********")
print(a.dropna(axis=1,how="any"))         ##屏蔽缺失数据的一行、
print("***********")
print(a.fillna(1))                        ##给缺失的部分加上数值
print("***********")
print(a.isnull())                          ##判断数据是否有缺失
print("***********")
```

```python
import pandas as pd
import numpy as np
a=pd.DataFrame(np.ones((3,4))*0,columns=["a","b","c","d"])
b=pd.DataFrame(np.ones((3,4))*1,columns=["a","b","c","d"])
c=pd.DataFrame(np.ones((3,4))*2,columns=["a","b","c","d"])
print(a)
print("*********")
print(b)
print("*********")
print(c)
print("*********") 
d=pd.concat([a,b,c],axis=0,ignore_index=True)                 ##上下合并，concat功能
print(d)
print("*********")
e=pd.concat([a,b,c],axis=1,ignore_index=True)
print(e)
print("*********")
```

```python
import pandas as pd
import numpy as np
a=pd.DataFrame(np.ones((3,4))*0,columns=["a","b","c","d"],index=[1,2,3])
b=pd.DataFrame(np.ones((3,4))*1,columns=["b","c","d","e"],index=[2,3,4])
print(a)
print("********")
print(b)
c=pd.concat([a,b],join="outer")                                ##outer与inner功能
print(c)
print("********")
d=pd.concat([a,b],join="inner",ignore_index=True)
print(d)
print("********")
e=pd.DataFrame(np.ones((3,4))*0,columns=["a","b","c","d"])
print(e)
print("********")
s=pd.Series([1,2,3,4],index=["a","b","c","d"])               ##appimend功能，添加某一行的数据
print(s)
print("********")
f=e.append(s,ignore_index=True)
print(f)
```

```python
import pandas as pd
import numpy as np
a=pd.DataFrame({"key":["K0","K1","K2","K3"],
               "A":["A0","A1","A2","A3"],
               "B":["B0","B1","B2","B3"]})
b=pd.DataFrame({"key":["K0","K1","K2","K3"],
               "C":["C0","C1","C2","C3"],
               "D":["D0","D1","D2","D3"]})
print(a)
print("*********")
print(b)
print("*********")
c=pd.merge(a,b,on="key")                        ##基于某一行合并,merge
print(c)
print("*********")
```

```python
import pandas as pd
import numpy as np
a=pd.DataFrame({"key1":["K0","K1","K2","K3"],
               "A":["A0","A1","A2","A3"],
               "B":["B0","B1","B2","B3"],
               "key2":["K0","K1","K2","K3"]})
b=pd.DataFrame({"key1":["K0","K1","K2","K3"],            ##有多个合并行，但合并行中数据完全相同
               "C":["C0","C1","C2","C3"],
               "D":["D0","D1","D2","D3"],
                "key2":["K0","K1","K2","K3"]})
print(a)
print("********")
print(b)
print("********")
c=pd.merge(a,b,on=["key1","key2"],how="inner")
print(c)
print("********")
```

```python
import pandas as pd
import numpy as np
a=pd.DataFrame({"key1":["K0","K0","K1","K2"],                    ##多个合并行，合并行不完全相同
               "A":["A0","A1","A2","A3"],
               "B":["B0","B1","B2","B3"],
               "key2":["K0","K1","K0","K1"]})
b=pd.DataFrame({"key1":["K0","K1","K1","K2"],
               "C":["C0","C1","C2","C3"],
               "D":["D0","D1","D2","D3"],
                "key2":["K0","K0","K0","K0"]})
print(a)
print("*****")
print(b)
print("*****")
c=pd.merge(a,b,on=["key1","key2"],how="inner")            ##merge,inner           
print(c)
print("*****")
d=pd.merge(a,b,on=["key1","key2"],how="outer")            ##merge,outer          
print(d)
print("*****")
e=pd.merge(a,b,on=["key1","key2"],how="left")            ##merge,left        
print(e)
print("*****")
f=pd.merge(a,b,on=["key1","key2"],how="right")            ##merge,right      
print(f)
```

```python
import pandas as pd
import numpy as np
a=pd.DataFrame({"coll":[0,1],"coll_left":["a","b"]})
b=pd.DataFrame({"coll":[1,1,2],"coll_right":[2,2,2]})
print(a)
print("*******")
print(b)
print("*******")
c=pd.merge(a,b,on="coll",how="outer",indicator=True)               ##indicator,判断合并类型
print(c)
```

```python
import pandas as pd
import numpy as np
a=pd.DataFrame({"A":["A0","A1","A2"],
               "B":["B0","B1","B2"]},
               index=["K0","K1","K2"])                                           ##index合并
b=pd.DataFrame({"C":["C0","C1","C2"],
               "D":["D0","D1","D2"]},
               index=["K0","K2","K3"])
print(a)
print("********")
print(b)
print("********")
print(pd.merge(a,b,left_index=True,right_index=True,how="inner"))
print("********")
print(pd.merge(a,b,left_index=True,right_index=True,how="outer"))
```

```python
import pandas as pd
import numpy as np
boys=pd.DataFrame({"k":["K0","K1","K2"],"age":[1,2,3]})
girls=pd.DataFrame({"k":["K0","K0","K3"],"age":[4,5,6]})
print(boys)
print("**********")
print(girls)
print("**********")
c=pd.merge(boys,girls,on="k",how="outer",suffixes=["_boys","_girls"])                ##index合并，并在columns加入名字
print(c)
```
